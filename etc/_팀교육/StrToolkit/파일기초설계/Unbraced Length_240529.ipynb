```python
PhiP = 0.9
PhiM = 0.9

Column_List = copy.deepcopy(Mat)

def Delete_NG_Element(List1) : 
    res = []
    for inner_list in List1 :
        if isinstance(inner_list[24], int) : 
            res.append(inner_list)
    return res

def Calculate_Dividing(List2) :
    for inner_list in List2 : 
        n = 1
        while True : 
            Lz = Ly/n
            Ag = inner_list[1]
            Syy = inner_list[8]
            Szz = inner_list[9]
            Zyy = inner_list[10]
            Zzz = inner_list[11]
            Izz = inner_list[13]
            ry = inner_list[15]
            rz = inner_list[16]
            J = inner_list[17]
            Cwp = inner_list[18]
            D = inner_list[20]
            W = inner_list[21]
            tw = inner_list[22]
            tf = inner_list[23]
            SR = max(K*Ly/ry, K*Lz/rz)
            Fe = (math.pi**2)*E/(SR**2)

            λrf = 0.56*math.sqrt(E/Fy)
            λrw = 1.49*math.sqrt(E/Fy)
            λf = W/(2*tf)
            λw = (D-2*tf)/tw
            
            if SR <= 4.71*math.sqrt(E/(Fy)) : 
                Fn = 0.658**(Fy/Fe)*Fy
            else : 
                Fn = 0.877*Fe
            

            if (λrf > λf) and (λrw > λw) :      # Flexural Buckling of Member without Slender Element
                ΦPn = PhiP*Fn*Ag

            elif (λrf < λf) and (λrw > λw) :    # Flexural Buckling of Member with Slender Flange
                Fel = 1.49*(λrf/λf)**2*Fy
                if λf <= λrf*math.sqrt(Fy/Fn) : 
                    Ae = Ag
                else : 
                    be = (W/2)*(1-0.22*math.sqrt(Fel/Fn)*math.sqrt(Fel/Fn))
                    Ae = Ag-(W*tf)+(2*be*tf)
                
                ΦPn = PhiP*Fn*Ae

            elif (λrf > λf) and (λrw < λw) :    # Flexural Buckling of Member with Slender Web
                Fel = 1.31*(λrw/λw)**2*Fy
                if λw <= λrw*math.sqrt(Fy/Fn) : 
                    Ae = Ag
                else : 
                    be = (D-2*tf)*(1-0.22*math.sqrt(Fel/Fn)*math.sqrt(Fel/Fn))
                    Ae = Ag-((D-2*tf)*tw)+(be*tw)

                ΦPn = PhiP*Fn*Ae

            else :                              # Flexural Buckling of Member with Slender Flange & Web
                Felf = 1.49*(λrf/λf)**2*Fy
                Felw = 1.31*(λrw/λw)**2*Fy
                if λf <= λrf*math.sqrt(Fy/Fn) and λw <= λrw*math.sqrt(Fy/Fn) : 
                    Ae = Ag
                elif λf > λrf*math.sqrt(Fy/Fn) and λw <= λrw*math.sqrt(Fy/Fn) : 
                    be = (W/2)*(1-0.22*math.sqrt(Felf/Fn)*math.sqrt(Felf/Fn))
                    Ae = Ag-(W*tf)+(2*be*tf)
                elif λf <= λrf*math.sqrt(Fy/Fn) and λw > λrw*math.sqrt(Fy/Fn) : 
                    be = (D-2*tf)*(1-0.22*math.sqrt(Felw/Fn)*math.sqrt(Felw/Fn))
                    Ae = Ag-((D-2*tf)*tw)+(be*tw)
                else : 
                    be = (W/2)*(1-0.22*math.sqrt(Felf/Fn)*math.sqrt(Felf/Fn))
                    he = (D-2*tf)*(1-0.22*math.sqrt(Felw/Fn)*math.sqrt(Felw/Fn))
                    Ae = (be*2*tf)+(he*tw)

                ΦPn = PhiP*Fn*Ae

            if λf < 0.38*math.sqrt(E/Fy) and λw < 3.76*math.sqrt(E/Fy) :                     # Lateral Torsional Buckling Major Axis(Compact Only)

                Cb = 1.0
                Mpy = Fy * Zyy
                Lb = Lz
                Lp = 1.76*rz*math.sqrt(E/Fy)
                Lr = 1.95*math.sqrt(math.sqrt(Izz*Cwp)/Syy)*E/(0.7*Fy)*math.sqrt(J/(Syy*(D-tf))+math.sqrt((J/(Syy*(D-tf)))**2+6.76*(0.7*Fy/E)**2))
                Fcr2 = Cb*math.pi**2*E/(Lb/math.sqrt(math.sqrt(Izz*Cwp)/Syy))**2*math.sqrt(1+0.078*J/(Syy*(D-tf))*(Lb/math.sqrt(math.sqrt(Izz*Cwp)/Syy))**2)

                if Lb <= Lp : 
                    Mny = Mpy
                elif Lp < Lb <= Lr : 
                    Mny = min(Cb*(Mpy-(Mpy-0.7*Fy*Syy)*(Lb-Lp)/(Lr-Lp)),Mpy)
                elif Lb > Lr : 
                    Mny = min(Fcr2*Syy,Mpy)

                ΦMny = PhiM*Mny

            Mnz = min(Fy*Zzz,1.6*Fy*Szz)                                                                 # Lateral Torsional Buckling Minor Axis (Compact Only)
            ΦMnz = PhiM*Mnz

            if ΦPn<Pu or ΦMny<Muy or ΦMnz<Muz or SR > 200:
                n = n + 1

            elif ΦPn>Pu and ΦMny>Muy and ΦMnz>Muz and SR < 200:
                n = n
                break

            if (ΦPn<Pu or ΦMny<Muy or ΦMnz<Muz or SR > 200) and n > 10 : 
                n = "NG"
                break

        inner_list.append(n)            # 24
        inner_list.append(ΦPn)          # 25
        inner_list.append(ΦMny)         # 26
        inner_list.append(ΦMnz)         # 27

    return List2

print(Column_List)

Calculate_Dividing(Column_List)

print(Column_List)

Selected_Column_List = Delete_NG_Element(Column_List)


print(Selected_Column_List)
```
