"34개의 단면적 데이터를 재사용하기 위해서 34개의 서로 다른 변수이름을 코드에 적어주는 것",

"34개의 단면적 데이터를 재사용하기 위해서 1개의 변수이름을 코드에 적어주는 것"<br>

시간 빌게이츠가 아니라면 당연히 후자를 통해 계산하는 것을 선택할 수 밖에 없습니다.

```python
단면적들 = [
    10, 10.5, 11, 11.5, 12, 12.5, 13, 13.5, 14, 14.5, 15, 15.5, 16, 16.5, 17, 17.5, 18, 18.5, 19, 19.5, 20, 20.5, 21, 21.5, 22, 22.5, 23, 23.5, 24, 24.5, 25, 25.5, 26, 26.5, 
]

길이들 = [
    6, 5, 6.5, 4, 4, 5, 5.5, 7, 7, 6, 5.5, 8, 5, 5, 5, 7, 7, 5, 6, 5.5, 6, 7, 7.5, 7, 7, 6, 5, 5, 5.5, 7, 7, 6, 5.5, 6.5, 
]
```

위와 같이 말입니다.

그리고

'단면적들' 리스트의 ***첫번째*** 항목과 '길이들' 리스트의 ***첫번째*** 항목을 곱해서 ***첫번째*** 부피를 만들고,<br>
'단면적들' 리스트의 ***두번째*** 항목과 '길이들' 리스트의 ***두번째*** 항목을 곱해서 ***두번째*** 부피를 만들고,<br>
...

34번째 항목까지 위의 패턴을 반복하면 원하는 결과를 얻을 수 있습니다.


```python
# 방법1

부피들1 = []
for n번째 in range(len(단면적들)):
    n번째_단면적 = 단면적들[n번째]
    n번째_길이 = 길이들[n번째]
    
    n번째_부피 = n번째_단면적 * n번째_길이

    부피들1.append(n번째_부피)
print("계산결과 :")
print(부피들1)
```

아까 전 안그래가<br>
" 첫번째_빔의_부피 = 첫번째_빔의_단면적 * 첫번째_빔의_길이 "<br>
...<br>
라고 작성했던 코드를 추상화 시켜서 반복하는 코드입니다.


'단면적들' 과 '길이들' 두가지 리스트의 각 원소에 가상으로 번호를 붙인뒤, 번호 순서대로 호출해가며 계산을 하는 코드로 이해하면 됩니다.

즉, 위의 표현식은 (예를 들어 빔이 총 5개라고 가정하면)<br><br>
"""<br>
