밖으로 뺄건 빼줍니다. 모듈 임포트 / 입력값 분리 등

```python
import math # 함수 내부에 모듈 임포트 선언이 존재하면, 함수를 호출할 때마다 임포트가 이루어지기 때문에 불필요한 오버헤드가 발생하므로, 바깥으로 빼주는 것이 합리적

def context1(num1,num2, mode = "계산모드1"):
    
    a = num1 + num2 #<---- 입력 추출 대응
    b = a * 2 + 11
    c_ = math.sin(b) + a**2 + b/2
    d = mode #<---- 입력 추출 대응
    if d == "계산모드1":
        c = c_ * math.e
    else:
        c = c_**2*math.e

    e = [1,2,b,11,c_/100,0,1]
    new_e = []
    for i in e:
        new_e.append(i * c/a)

    cnt = 0
    for i in new_e:
        if i < 5:
            cnt += 1

    result = e[:cnt+1]


print(context1(1,2)) # 현재는 함수를 만들어 내부에서 연산을 수행해도, 호출했을 때 결과가 없는 함수
```

함수 내부의 연산 결과를 외부로 반환해주는 return 예약어를 사용합니다.<br>
여기까지만 리팩토링해도, 이제 다른 사람과 협업이 가능한 코드가 됩니다.<br>
내부의 코드 정리는 차차 해나가도 됩니다.

```python
import math

def context1(num1,num2, mode = "계산모드1"):
    
    a = num1 + num2
    b = a * 2 + 11
    c_ = math.sin(b) + a**2 + b/2
    d = mode
    if d == "계산모드1":
        c = c_ * math.e
    else:
        c = c_**2*math.e

    e = [1,2,b,11,c_/100,0,1]
    new_e = []
    for i in e:
        new_e.append(i * c/a)

    cnt = 0
    for i in new_e:
        if i < 5:
            cnt += 1

    result = e[:cnt+1]
    return result


print(context1(1,2))
```

함수 내부의 알고리즘을 좀더 세부단위로 주석을 달아봅니다.

```python
def context1(num1,num2, mode = "계산모드1"):
    
    a = num1 + num2 # 입력값을 받아서 새로운 a값 연산
    b = a * 2 + 11 # a를 활용한 새로운 b값 연산
    c_ = math.sin(b) + a**2 + b/2 # a와 b를 활용한 새로운 c_값 연산
    d = mode # 입력받은 계산모드 결정 용 문자열
    if d == "계산모드1":
        c = c_ * math.e # 계산모드에 따라 c_를 활용해 연산된 새로운 c값
    else:
        c = c_**2*math.e # 계산모드에 따라 c_를 활용해 연산된 새로운 c값

    e = [1,2,b,11,c_/100,0,1] # b, c 를 활용한 새로운 리스트 연산
    new_e = []
    for i in e:
        new_e.append(i * c/a) # 리스트e를 활용한 새로운 new_e 리스트 연산

    cnt = 0
    for i in new_e:
        if i < 5:
            cnt += 1 # 연산된 new_e 리스트의 원소 중 5보다 작은 것의 

    result = e[:cnt+1]
    return result
```

```python
입력값1 = 1
입력값2 = 2
입력값3 = 10
또다른어떤입력값 = "계산모드1"
입력값4 = 7
```
